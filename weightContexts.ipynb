These functions assume that you have the following datasets produced with `semasioFlow` (on Python):

- A dataframe with one row per context word and columns with frequency and weight information (most importantly, a column with the weight you would use for first-order filtering/weighting). The context word ID column is called `cw`. The file would be typically called '{lemma}.ppmi.tsv' and stored with your github stuff (for the viz).
- A dataframe with one row per token and columns with annotation values and such, as well as columns with semicolon-separated context words. The token ID column is called `_id` and the context word columns start with `_cws.` followed by the (first part of) the name of the model. The file would be typically called `{lemma}.variables.tsv` and live in your github directory for the visualization.
- A dataframe with one row per token per context word, with corpus-based information. This is generated by `semasioFlow.listContextwords()` and will at least have a `cw` column with the type of the context word, a `target_lemma` column with the type of the target, a `word` column with the word form, `token_id` and `cw_id` column, as well as distance between context word and target and stuff like that. It will be typically called `{lemma}.cws.detail.tsv`

The second dataframe must be loaded separately, while the first and third dataframes will be loaded and properly merged by `semcloud::setupConcordancer()`.
There are two ways to point to the files in this function. If their names correspond to the naming convention _and_ are stored in the same directory `input_dir`, you can run `setupConcordancer(lemma, input_dir)`. Otherwise, you can use `setupConcordancer(cws_detail_path = 'path/to/cws.detail.tsv', ppmi_path = 'path/to/ppmi.tsv')`.

## Build weighted concordances

Then you use both dataframes and the name of your lemma for the Ultimate Function, `semcloud::weightConcordance()`, which will return an enriched `variables` dataframe with a raw context in the `_ctxt.raw` and weighted contexts in columns starting with `_ctxt.`, followed by (the first part of) the name of the model.

The default settings of these functions assume a certain pattern in the names of the columns and how to extract parameter settings from them, but they can be adapted with _ad hoc_ functions as arguments. For example, the value of `sup_weight_fun` is a function that takes a string with weighting information (in my case 'PPMIweight', 'PPMIselection' or 'PPMIno') and returns a boolean value indicating whether PPMI values should be included as superindices in the weighted concordance. By default, it does when the string ends with 'weight'; if you wanted PPMI values to _always_ be included next to captured words, you could run `weightConcordance(variables, cws, lemma, sup_weight_fun = function(x) return(TRUE))`.

```r
lemmas <- setdiff(list.files(github_dir), c('euclidean_register.tsv', 'lemmarel', 'temp', 'zwart', 'church', 'destroy'))
length(lemmas)
```

```r
for (lemma in lemmas) {
    cws <- setupConcordancer(cws_detail_path = file.path(cws_detail_dir, paste0(lemma, '.cws.detail.tsv')),
                         ppmi_path = file.path(github_dir, lemma, paste0(lemma, '.ppmi.tsv')))
    variables <- readr::read_tsv(file.path(github_dir, lemma, paste0(lemma, ('.variables.tsv'))),
                            show_col_types = F, lazy = F)
    ctxts <- weightConcordance(variables, cws, lemma)
    write_tsv(ctxts, file.path(github_dir, lemma, paste0(lemma, ('.variables.tsv'))))
}
```
